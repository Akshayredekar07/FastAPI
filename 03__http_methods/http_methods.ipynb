### **Website Architecture and Interactions**

#### **Website Architecture**
- **Definition**: A website is a special type of software installed on a server, accessed by clients (e.g., browsers) via two-way communication using HTTP protocols.
- **Components**:
  - **Server**: The machine hosting the website, processing requests, and serving data.
  - **Client**: The device or software (e.g., browser) accessing the website.
  - **Communication**: Occurs via HTTP protocols, enabling data exchange between client and server.

#### **Types of Websites**
- **Static Websites**:
  - Low interaction, primarily for displaying information.
  - Examples: Government websites, blogs.
  - Limited user actions, mostly read-only (viewing content).
- **Dynamic Websites**:
  - Support extensive interactions, enabling data manipulation.
  - Examples: Zomato, account management systems.
  - Facilitate four primary interaction types (CRUD).

#### **Four Types of Interactions on Dynamic Websites (CRUD)**
1. **Create**:
   - Action: Adding new data.
   - Example: Placing an order on Zomato, submitting a form.
   - HTTP Method: POST request.
2. **Read**:
   - Action: Retrieving or viewing data.
   - Example: Browsing menu items, viewing a user profile.
   - HTTP Method: GET request.
3. **Update**:
   - Action: Modifying existing data.
   - Example: Changing a username, updating a delivery address.
   - HTTP Method: PUT or PATCH request.
4. **Delete**:
   - Action: Removing data.
   - Example: Canceling an order, deleting a comment.
   - HTTP Method: DELETE request.

#### **Additional Notes**
- **CRUD in Context**: These four interactions (Create, Read, Update, Delete) are the primary ways users engage with dynamic websites, supported by client-server communication over HTTP.
- **Clarification**: Static websites primarily support Read interactions, while dynamic websites enable all CRUD operations for enhanced functionality.
