### **RECAP**

1. **Client-Server Interaction**
    - The client sends a **POST** request to the server.
    - The request body contains **JSON** data.

2. **Validation**
    - The data is **validated** using a **Pydantic model**.
    - Example:  
      `"thirty"` → `30` 
      Which is not valid data format

3. **Data Storage**
    - The validated data is added as a **new record**
